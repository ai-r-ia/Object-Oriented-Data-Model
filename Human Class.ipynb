```python
import numpy as np
import random
import sys
```

```python
class human:
    bodycal=10
    money=100
    reserved=10
    isalive=True
    daycount=0
    oneMeal=10
    hours=0
    action=0
    
    def __init__(self,reserved=10,money=100,bodycal=10):
        self.reserved=reserved
        self.money=money
        self.bodycal=bodycal
    def timeofdeath(self,x):
        if x==1:
            print("Died while exercising")
        elif x==3:
            print("Died while working")
        elif x==5:
            print("Died in sleep")
                      
    def bodycalories(self,x):
        if self.bodycal>0:
            if x<0:
                self.bodycal=self.bodycal-(x/100)*self.bodycal
            else:
                self.bodycal=self.bodycal+(x/100)*self.bodycal
        else:
            print("You're dead")
            self.isalive=False
            self.timeofdeath(self.action)
            print("lived for",self.daycount, "days")
            sys.exit()

    def exercise(self):
        if random.getrandbits(1)==1:
            self.bodycal-=4
            self.hours+=2
            self.action=1
            print("exercised")
            
    def eat(self):
        if self.oneMeal>self.reserved:
            print("came to eat, but going to shop")
            self.shop(self.oneMeal-self.reserved)
        else:
            self.bodycalories(+0.4)
            self.hours+=2
            self.action=2
            print("ate")
    def work(self):
        self.bodycal-=0.2
        self.money+=5
        self.hours+=8
        self.action=3
        print("worked")
    def shop(self,x):
        if self.money>x and self.money>=20:           
            self.money-=20
            self.reserved+=20
            print("shopped when rich")
        elif self.money>=x and self.money<20:
            self.money-=x
            self.reserved+=x
            print(self.money,"shopped when poor")
        elif self.money<=x and self.money>=5:
            self.money-=5
            self.reserved+=5
            print("chocolate")
        else:
            print(self.money, "NO MONEY")
        self.hours+=3
        self.action=4
#         print("shopped")
    def sleep(self):
        self.bodycalories(-0.7)
        self.hours+=8
        if self.daycount<100:
            self.daycount+=1
        else:
            print("Reached age limit of 100 yrs")
            sys.exit
        self.action=5
        print("slept")
#         self.isalive=False
    def life(self): 
        while (self.isalive): 
            self.exercise()
            self.eat()
            self.work()
            self.shop(5)
            self.eat()
            self.sleep()
#         print(self.hours, self.daycount)
```
